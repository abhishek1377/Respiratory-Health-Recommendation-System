```python
import pandas as pd
import numpy as np
```

```python
merged_dataset = pd.read_csv("merged_pollutant_dataset.csv")
neighbors = pd.read_csv("neighbors.csv")
```

```python
# merged_dataset["date"]
# merged_dataset['date_state_code_county_code'] = merged_dataset["date"] + merged_dataset["state_code"] + \
# merged_dataset["count_code"]

new_df = merged_dataset.copy()
new_df["date_state_code_county_code"] = merged_dataset["date"].apply(str) + "_"+ merged_dataset["state_code"].apply(str) + \
                                        "_" + merged_dataset["county_code"].apply(str)

new_df.index = new_df["date_state_code_county_code"]
```

```python
new_df = new_df.drop(columns = ["date_state_code_county_code"])
```

```python
new_df
```

```python
#creating new columns and adding them to working dataset

columns_neighbors = []

for k in range(1,6):
    for i in merged_dataset.columns:
        columns_neighbors.append("neighbor"+str(k)+"_"+i)
 

working_df = merged_dataset.copy()

for column in columns_neighbors:
    working_df[column] = np.nan
    
```

```python
#neighbor1

neighbor1_columns = list(working_df.columns)[20:40]
neighbor1_df = pd.DataFrame(columns = neighbor1_columns)
neighbor1_list = []

for i in range(len(merged_dataset)):

    date = merged_dataset["date"][i]
    state = merged_dataset["state_code"][i]
    county = merged_dataset["county_code"][i]
#     print(i)
    neighbors_row = neighbors[(neighbors["state_code"] == state) & (neighbors["county_code"] == county)].to_numpy()[0]
#     print(neighbors_row)
    neighbor_state = neighbors_row[2]
    neighbor_county = neighbors_row[3]
    neighbor_info = new_df.loc[str(date)+"_"+str(neighbor_state)+"_"+str(neighbor_county)]
#     print(neighbor_info)
    #         working_df.iloc[i:i+1,10*j:10*j+20] = neighbor_info
    neighbor1_list.append(neighbor_info)
#     display(neighbor1_df)
    
np_a = np.asarray(neighbor1_list)
np_a
neighbor1_df = pd.DataFrame(data = np_a, columns = neighbor1_columns)
# working_df
neighbor1_df
```
