```python
merged_users_forecasts_df["max_AQI_pollutant"].replace("pm25nfrm", "pm25", inplace = True)
# merged_users_forecasts_df
```

```python
merged_users_forecasts_df["max_AQI_pollutant"].replace("pm25frm", "pm25", inplace = True)
# merged_users_forecasts_df
```

```python
aqi_reco.head()
```

```python
merged_users_forecasts_reco_df_new = merged_users_forecasts_df.merge(aqi_reco, left_on = ["max_AQI_pollutant"],  right_on = ["Worst_Pollutant"])
```

```python
mask = (merged_users_forecasts_reco_df_new["max_pollutant_AQI"] >= merged_users_forecasts_reco_df_new["Min_AQI"]) \
        & (merged_users_forecasts_reco_df_new["max_pollutant_AQI"] <= merged_users_forecasts_reco_df_new["Max_AQI"] )
```

```python
final_recos_df = merged_users_forecasts_reco_df_new[mask]
```

```python
final_recos_df = final_recos_df.reset_index(drop = True)
final_recos_df
```
