```python
def get_data(merged_pollutant_dataset_with_neighbors_sorted_pairs,pair_number):
    state_county_df = merged_pollutant_dataset_with_neighbors_sorted_pairs.iloc[4155*pair_number:4155*(pair_number+1),:]
    return state_county_df
```

```python
def data_cleaning(state_county_df):

    # dropping columns ones with all null values
    state_county_df_reduced = state_county_df.dropna(axis=1, how='all')
    
    # dropping columns with all same/equal values
    nunique = state_county_df_reduced.nunique()
    cols_to_drop = nunique[nunique == 1].index
    state_county_df_reduced = state_county_df_reduced.drop(cols_to_drop, axis=1)

    # dropping columns with date types
    data_types = state_county_df_reduced.dtypes
    cols_to_drop = data_types[data_types == "object"].index
    state_county_df_reduced = state_county_df_reduced.drop(cols_to_drop, axis=1)
    
    return state_county_df_reduced
```

```python
def cleaned_data_plots(state_county_df_reduced):

    fig, axes = plt.subplots(nrows=4, ncols=3, dpi=120, figsize=(10,8))
    for i, ax in enumerate(axes.flatten()):
        data = state_county_df_reduced[state_county_df_reduced.columns[i]]
        ax.plot(data, color="violet", linewidth=1)
        # Decorations
        ax.set_title(state_county_df_reduced.columns[i])
        ax.xaxis.set_ticks_position('none')
        ax.yaxis.set_ticks_position('none')
        ax.spines["top"].set_alpha(0)
        ax.tick_params(labelsize=6)

    plt.tight_layout();
```

```python
# Interpolation for missing values
def interpolation(state_county_df_reduced, interpolation_method = 'time'):
    
    state_county_df_reduced_interpolated = state_county_df_reduced.interpolate(method = interpolation_method)
    
    return state_county_df_reduced_interpolated
```

```python
# Selecting columns with all non-null values after interpolation 

def get_all_non_null_columns(state_county_df_reduced_interpolated):

    state_county_non_null_df = state_county_df_reduced_interpolated.dropna(axis=1, how='any')
    
    return state_county_non_null_df
```

```python
def split_data(state_county_non_null_df, split_ratio = 0.08):

    split_time = int(4155*split_ratio)
    train = state_county_non_null_df.iloc[:len(state_county_non_null_df)-split_time]
    test = state_county_non_null_df.iloc[len(state_county_non_null_df)-split_time:] 
    
    return (train, test, split_time)
```

```python
# Training the VAR model
```

```python
def model_train(train):

    model = VAR(train) 
    
    return model
```

```python
#testing for the best lag value

def best_lag(model):
    results_summary = model.select_order(30).summary()
    file = results_summary.as_html()
    results_summary_df = pd.read_html(file, header=0, index_col=0)[0]
    results_summary_df.dtypes
    
    # Getting optimum lag

    aics = list(results_summary_df["AIC"].values)
    aic_min = 7
    for i in range(30):
        if '*' in aics[i]:
    #         print(i)
            aic_min = i
            break

    bics = list(results_summary_df["BIC"].values)
    bic_min = 7
    for i in range(30):
        if '*' in aics[i]:
    #         print(i)
            bic_min = i
            break


    fpes = list(results_summary_df["FPE"].values)
    fpe_min = 7
    for i in range(30):
        if '*' in fpes[i]:
    #         print(i)
            fpe_min = i
            break

    optimum_lag = max(45, aic_min*3,bic_min*3,fpe_min*3)
    return optimum_lag
```

```python
#Fitting the model with the best lag value

def model_fit(model,optimum_lag):

    results = model.fit(optimum_lag)
    return results
```

```python
#getting long-term forecasts using the model

def long_term_forecast(results,train,test, split_time):
    lag = results.k_ar
    forecasts = results.forecast(train.values[-lag:], steps = split_time)
    forecasts_df = pd.DataFrame(forecasts, columns = list(test.columns))
    return forecasts_df
    
```

```python
# getting plots of actual test data vs forecasts

def forecasts_vs_actual_plot(forecasts_df, test, key):

    forecasts_df.index = test.index
    # display(forecasts_df)
    fig, axes = plt.subplots(nrows= 4, ncols=2, dpi=150, figsize=(13,13))
    for i, (col,ax) in enumerate(zip(forecasts_df.columns, axes.flatten())):
    #     print(forecasts_df[col], test[col])
        test[col][:70].plot(legend=True, ax=ax, label = "Actual");
        forecasts_df[col][:70].plot(legend=True, ax=ax,label = "Forecasts")  
        ax.set_title(key +"_"+ col + ": Forecast vs Actuals")
        ax.xaxis.set_ticks_position('none')
        ax.yaxis.set_ticks_position('none')
        ax.spines["top"].set_alpha(0)
        ax.tick_params(labelsize=6)

    plt.tight_layout();


# As we can see.. forecasts tend to converge to a fix value after certain time periods. Long time forecasting is unreliable.
```

```python
# creating functions to evaluate the models

def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    return({'mape':mape, 'me':me, 'mae': mae, 
            'mpe': mpe, 'rmse':rmse},["mape", "me" , "mae" , "mpe" , "rmse"], [mape, me, mae, mpe, rmse])


def evaluation_metric(forcast_df, actual_df):
    evaluations_forcast_df = {}
    results_forcast_df = None
    
#     print("forecasts:", forcast_df)
    
#     print("test", actual_df)

    for i in range(min(8,len(forcast_df.columns))):
        results_forcast_df = forecast_accuracy(forcast_df[forcast_df.columns[i]].values, actual_df[actual_df.columns[i]].values)
        evaluations_forcast_df[forcast_df.columns[i]] = results_forcast_df[2]

    evaluations_df = pd.DataFrame.from_dict(evaluations_forcast_df, orient='index',
                       columns= results_forcast_df[1])

    return evaluations_df

```

```python
# Checking how VAR estimates the training data

def var_training_data_performace(train, results):
    lag = results.k_ar
    estimations = np.empty(train.iloc[:len(train)-lag].shape)

    for i in range(len(train)-lag):
        estimate = results.forecast(train.values[i:i+lag], steps = 1)
        estimations[i] = estimate
 
    estimations_df = pd.DataFrame(estimations, columns = list(train.columns))
    train_lag = train.iloc[lag:]
    estimations_df.index = train_lag.index
    
    return estimations_df
```

```python

def estimates_vs_actual_plot(estimations_df,train, test,lag,key):
    
    train_lag = train.iloc[lag:]
    fig, axes = plt.subplots(nrows=4, ncols=2, dpi=150, figsize=(13,13))
    for i, (col,ax) in enumerate(zip(estimations_df.columns, axes.flatten())):
    #     print(forecasts_df[col], test[col])
        train_lag[col][:70].plot(legend=True, ax=ax, label = "Actual");
        estimations_df[col][:70].plot(legend=True, ax=ax, label = "Estimates")

        ax.set_title(key +"_"+ col + ": Estimates vs Actuals")
        ax.xaxis.set_ticks_position('none')
        ax.yaxis.set_ticks_position('none')
        ax.spines["top"].set_alpha(0)
        ax.tick_params(labelsize=6)

    plt.tight_layout();
```

```python
# Checking how VAR predicts the test data(one day at a time)

def var_test_data_performance_one_day_at_a_time(state_county_non_null_df,results,test,split_time):
    lag = results.k_ar
    predictors = state_county_non_null_df.iloc[-split_time-lag:]
    predictions = np.empty(test.shape)
    for i in range(test.shape[0]):
        prediction = results.forecast(predictors.values[i:i+lag], steps = 1)
        predictions[i] = prediction

    predictions_df = pd.DataFrame(predictions, columns = list(test.columns))
    predictions_df.index = test.index
    
    return predictions_df
```

```python
def test_prediction_vs_actual_plot(predictions_df,test, key):

    fig, axes = plt.subplots(nrows=4, ncols=2, dpi=150, figsize=(13,13))
    for i, (col,ax) in enumerate(zip(predictions_df.columns, axes.flatten())):
    #     print(forecasts_df[col], test[col])
        test[col][:70].plot(legend=True, ax=ax, label = "Actual");
        predictions_df[col][:70].plot(legend=True, ax=ax, label = "Predictions")

        ax.set_title(key +"_"+ col + ": Predictions vs Actuals")
        ax.xaxis.set_ticks_position('none')
        ax.yaxis.set_ticks_position('none')
        ax.spines["top"].set_alpha(0)
        ax.tick_params(labelsize=6)

    plt.tight_layout();
```
